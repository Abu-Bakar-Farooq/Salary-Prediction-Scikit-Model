```python
from sklearn.linear_model import Ridge
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np

# Assuming x_train is a DataFrame with mixed data types
# First, identify categorical columns (those with string values)
categorical_cols = X_train.select_dtypes(include=['object']).columns

# Create a copy of the data to avoid modifying the original
X_train_processed = X_train.copy()

# Apply one-hot encoding to categorical columns
if len(categorical_cols) > 0:
    # Option 1: Using pandas get_dummies (simpler approach)
    X_train_processed = pd.get_dummies(X_train_processed, columns=categorical_cols, drop_first=True)
    
    # Alternative Option 2: Using OneHotEncoder from sklearn
    # encoder = OneHotEncoder(sparse=False, drop='first')
    # encoded_data = encoder.fit_transform(x_train[categorical_cols])
    # encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_cols))
    # x_train_processed = x_train_processed.drop(categorical_cols, axis=1)
    # x_train_processed = pd.concat([x_train_processed, encoded_df], axis=1)

# Now fit the Ridge model with the processed data
ridge = Ridge()
ridge.fit(X_train_processed, y_train)
```

```python
ridge_pred = ridge.predict(X_test)
print("Ridge error: ", mean_squared_error(y_test, ridge_pred))
print("Ridge score:", r2_score(y_test, ridge_pred))
```
